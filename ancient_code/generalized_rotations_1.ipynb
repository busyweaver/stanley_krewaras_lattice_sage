```python
def derniere_feuille_gauche(r):
    if r[0]==0:
        return r
    else:
        return derniere_feuille_gauche(r[0])
    
def derniere_feuille_droite(r):
    if r[len(r)-1]==0:
        return r
    else:
        return derniere_feuille_droite(r[len(r)-1])
def dernier_noeud(r,pere,d):
    if r==0:
        return [r,pere]
    else:
        return dernier_noeud(r[d-1],r,d)
        
def dernier_noeud_droite(r,pere,d):
    #d'abord gauche puis tout Ã  droite
    if r[d-1]==0:
        return [r,pere]
    else:
        return dernier_noeud_droite(r[d-1],r,d)
def creer_premiere(n,dire,d):
    if n%(d-1)!=1 and d>2:
        print ("creer premier, probleme taille")
        return -1
    r=[0 for i in range(d)]
    n=n-(d)
    while(n>0):
        if dire==1:
            ll=derniere_feuille_gauche(r)
            ll[0]=[0 for i in range(d)]
        else:
            ll=derniere_feuille_droite(r)
            ll[len(ll)-1]=[0 for i in range(d)]
        n=n-(d-1)
    return r
# creer_premiere(7,0,3)
# def nb_parentheses(l):
#     if l==0:
#         return 0
#     elif l==[0,0,0]:
#         return 1
#     else:
#         return 1+nb_parentheses(l[0])+nb_parentheses(l[1])+nb_parentheses(l[2])
```

```python
def tup(l):
    if l==0:
        return 0
    if l==[0,0,0]:
        return (0,0,0)
    else:
        ll=[]
        for e in l:
            ll.append(tup(e))
        return tuple(ll)      
```

```python
def generer_tout_depuis(c,typ,dire):
    d=len(c)
    D=dict()
    i=0
    D[tup(c)]=i
    pile=[c]
    res=[c]
    poset=[]
    while(pile!=[]):
        x=pile[0]
        #print "verif",x,res
        if  1:
            #res.append(x)
            pile.pop(0)
            #print "hey",x
            if typ==0:
                if dire==0:
                    l=rotations_tamari_g(x,d)
                else:
                    l=rotations_tamari_d(x,d)
            else:
                if dire==0:
                    l=rotations_semi_g(x,d)
                else:
                    l=rotations_semi_d(x,d)
            elem=[]
            for e in l:
                if e not in res:
                    res.append(e)
                    i=i+1
                    D[tup(e)]=i
                    elem.append(D[tup(e)])
                    pile.append(e)
                else:
                    elem.append(D[tup(e)])
            poset.append(elem)
        else:
            pile.pop(0)
    return [res,poset]

#generer renvoie [1,2] 1:  cest les elements 2:poset
def generer_toutes_expressions(n,typ,dire,d):
    if n%(d-1)!=1 and d>2:
        return -1
    c=creer_premiere(n,dire,d)
    #print "premier",c
    return generer_tout_depuis(c,typ,dire)

def nb_elem_expre(n,typ,d):
    if n%(d-1)!=1 and d>2:
        return -1
    l=generer_toutes_expressions(n,typ,dire,d)[0]
    return len(l)
# l=generer_toutes_expressions(7,0,0,3)
# len(l[0])
# for e in l[0]:
#     print e
# l=generer_toutes_expressions(5,0)
# print l[0]
# for e in l[1]:
#     print e
# print len(l[0])
#generer_tout_depuis([[[0, 0, 0], 0, 0], 0, 0],0)
# for e in generer_toutes_expressions(7,0)[0]:
#     print e
# print "hi"
# for e in generer_toutes_expressions(7,1)[0]:
#     print e
```

```python
def bijection_sommet(l1,l2):
    res=[]
    for i in srange(0,len(l1)):
        res.append( l2.index(l1[i]))
    return res   
#print generer_toutes_expressions(n,0,3)[1]
```

```python
def verif_est_plus_petit_brute(x,y,typ,dire):
    l=generer_tout_depuis(x,typ)
    l=l[0]
    if y in l:
        return 1
    else:
        return 0
```

```python
def generate_pairs(n,limit=Infinity):
    l=[]
    lim=0
    for i in srange(0,n):
        for j in srange(0,n):
            r=[i,j]
            if r not in l:
                lim=lim+1
                l.append(r)
                if lim>limit:
                    return l
    return l
def intervals(n,typ):
    arete=[]
    l=generer_toutes_expressions(n,typ,dire)
    p=generate_pairs(len(l))
    for [i,j] in p:
        if verif_est_plus_petit_brute(l[i],l[j]):
            arete.append([i,j])
    return arete
```

```python
def intervals_intersection(n,dire,d):
    arete=[]
    l=generer_toutes_expressions(n,1,dire,d)
    l=l[0]
    p=generate_pairs(len(l))
    for [i,j] in p:
        if verif_est_plus_petit_brute(l[i],l[j],0) and verif_est_plus_petit_brute(l[i],l[j],1):
            arete.append([i,j])
    return arete

def intervals_imply(n,d):
    arete=[]
    l=generer_toutes_expressions(n,1,d)
    l=l[0]
    nbp=0
    nb=0
    p=generate_pairs(len(l))
    for [i,j] in p:
        if verif_est_plus_petit_brute(l[i],l[j],1):
            nb=nb+1
            if verif_est_plus_petit_brute(l[i],l[j],0)==0:
                nbp=nbp+1
    return nbp/nb
#proportion pour taille 9 si tamari est vrai alors semi vrai = 5/19 l'inverse 33/107
#print (intervals_imply(7,3) )
```
